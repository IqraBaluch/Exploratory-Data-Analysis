# EDA on IPL Data

Data exploration technique to understand the various aspects of the data.
1. Understand the Data
2. Clean the Data
3. Analyis of Relationship between variables

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
ipl = pd.read_csv('OneDrive\Desktop/matches.csv')
ipl
```

```python
ipl.isnull().sum()
```

```python
ipl_data = ipl.drop(['dl_applied','umpire3'], axis=1)
ipl_data.head()
```

```python
corelation = ipl_data.corr()
```

```python
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns, annot = True )
```

```python
sns.pairplot(ipl_data, hue = "season")
```

```python
sns.catplot(x='batting_', kind = 'box', data= ipl_data)
```

```python
# In barPlot 1st parameter takes categoical value, 2nd parameter takes numerical values

plt.figure(figsize=(3,2))
plt.bar(list(ipl['player_of_match'].value_counts()[0:3].keys()),list(ipl['player_of_match'].value_counts()[0:3]),color='purple')
plt.show()
```

```python
batting_first = ipl_data[ipl_data['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
batting_first.head()
```

```python
batting_first.tail()
```

```python
plt.figure(figsize=(5,4))
plt.hist(batting_first['win_by_runs'])
plt.title('Distriution of Runs')
plt.xlabel("Runs")
plt.show()
```

```python
# batting first winner
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]),color=['blue','green','red'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
batting_second = ipl_data[ipl_data['win_by_wickets']!=0]
batting_second.head()
```
